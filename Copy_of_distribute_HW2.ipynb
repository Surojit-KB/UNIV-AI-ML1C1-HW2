#### 1.1.2 How has the popularity of these 30 genres changed with time?

Now that we have the top 30 genres, we can understand the trajectory of these various genres with time.

**Hint:**<br>
To do this, create a subframe of the ranking and year for each genre.<br>
Use the the groupby() function to group by year. <br>
This will give you a dataframe that contains the rankings of every song from that genre in a given year. 

Next, create a dictionary `genreinfo` keyed by genre, whose value is a tuple of 3 Pandas series: 
- the mean of the rankings per year for that genre, 
- the standard deviation on those rankings per year for that genre,
- the series groupby object for the rankings per year created from before

Here is an example of what one dictionary for one particular genre should look like.
This is the output of `genreinfo['/wiki/Hip_Hop']`
```
{'meanseries': year
 1992    45.222222
 1993    41.843750
 1994    48.400000
 ..
 2019    49.617284
 2020    50.152941
 'rankingseries': <pandas.core.groupby.generic.SeriesGroupBy object at 0x7f546e294a90>,
 'stdseries': year
 1992    28.544612
 1993    26.908578
 1994    28.230355
 ....
 2019    27.151689
 2020    28.295846}
 ```

````python
#your code here
genreinfo={}
for genre in top_genres.index:#Accesing each individual genre from the top genre list
  genre_series=largedf[largedf[genre]==1][['ranking','year']].groupby('year')#Creating a group by dataframe
  genre_mean=largedf[largedf[genre]==1][['ranking','year']].groupby('year').mean()#Calculating mean for each genre
  genre_std=largedf[largedf[genre]==1][['ranking','year']].groupby('year').std()#Calculation standard deviation for the each genre
  genreinfo[genre]={'meanseries':genre_mean,
                    'rankingseries':genre_series,
                    'stdseries':genre_std}#Creating a dictionary entry
````

#### 1.2.2 What if we used a different metric?

A singer should be scored higher if he or she appears higher in the rankings. So we'd say that a singer who appeared once at a higher and once at a lower ranking is a "higher quality" singer than one who appeared twice at a lower ranking. 

Think of a metric to do the above and plot another bar chart to determine the 'highest quality' singer.

````python
#Create a metric 

#your code here
#Using new metric 1.
#New metric was chosen as the mean ranking for the singers with minimum 15 appearance
best_singers=largedf[largedf['band_singer'].isin(data.index)].groupby('band_singer')['ranking'].mean().sort_values()#Getting the singers from the above data and obtaining there mean rank
fig,ax=plt.subplots(figsize=(20,10))
ax=sns.barplot(x=best_singers.index,y=best_singers.values)#Barplot for the mean ranking
ax.set_xticklabels(labels=[singer for singer in best_singers.index],fontdict={'fontsize':12},rotation='vertical');
ax.set_ylabel('Mean ranking')
ax.set_xlabel('Artist/Singer')
ax.set_title("Mean ranking of most occuring artists in Billboard's Top 100 list");
````

````python
#Creating a merged dataframe to showcase all the information obtained
best_merge=pd.merge(best_singers,data,left_on=best_singers.index,right_on=data.index).rename(columns={'band_singer':'count','key_0':'Artist/band'})
#New metric. To give importance to the number of appearances, the mean was divided by the number of appearance.
#More discussion on this metric later
best_merge['Adjusted rank']=best_merge['ranking']/best_merge['count']
best_merge=best_merge.sort_values('Adjusted rank')
best_merge
````

````python
#Plot for the new metric
fig,ax=plt.subplots(figsize=(20,10))
ax=sns.barplot(x=best_merge['Artist/band'],y=best_merge['Adjusted rank'])
ax.set_ylabel('Adjusted ranking')
ax.set_xlabel('Artist/Singer')
plt.xticks(rotation=90,fontsize=15)
ax.set_title("Adjusted ranking of most occuring artists in Billboard's Top 100 list");
````

####2.1.2 Request the Billboard Weekly pages for 2020

Billboard releases its weekly ratings on saturdays. 
Billboard makes the lists for every saturday available at:
`http://www.billboard.com/charts/hot-100/saturday`, for example, http://www.billboard.com/charts/hot-100/2014-12-27.

Therefore, before extracting data from the webpages, we need to first find all the saturdats for a given year. 

We have written a function `allsats(year, fs)` to get all the saturdays in a given year, and given the day part of the date of the first saturday of the year in january.

````python
from datetime import date, timedelta
#getting all sats
def allsats(year, fs):
    days=[]
    d = date(year, 1, fs)       
    while d.year == year:
        days.append(d.strftime("%Y-%m-%d"))
        d += timedelta(days = 7)
    return days
````

The first Saturday in 2020 is Jan 4th:

````python
#calling function allsats
sats=allsats(2020, 4)
len(sats)
````

Now, we need to scrape the webpage using pythons `request` module. 

Store the text from your requests in a dictionary called `satdict`. This dictionary should have as its keys keys the saturday date strings seen above in the list `sats` , and one additional key, `end_year`. ### 2.2 Parsing and Pre-Processing
[20 points]

#### 2.2.1 Parsing the HTML text

Now that you have the HTML text for both the year-end and weekly results, we can parse the text to extract the information we need. 

To do this, you will need to write 2 functions:

1. A function `get_weekly(weekstring, indict)`.<br>
  * It should take in two parameters: 
    * `weekstring`: the date of the corresponding saturday (For example: '2020-01-04')
    * `indict`: a dictionary with *keys* as the saturdays and *values* as the text of the billboard page for that day.
  *  It should return a **list** of tuples `(ranking, songdict)`where each tuple corresponds to one song in that week's top-100. 
    * `ranking` is the ranking of that song
    * `songdict` is a dictionary. The dictionary `songdict` should look like this: <br>
    `songdict=dict(ranking=ranking, title=title)`
    <br>where `title` is the title of the song and `ranking` is the same as the first element of the tuple.
    <br><br>



2. A function `get_year(weekstring,indict)` which extracts the same information as above for the 2020 year-end results. 
  * It should take in two parameters: 
    * `weekstring`: the `end_year` key, and 
    * `indict`: value from the dictionary `satdict`

````python
"""
Function
--------
get_weekly

Inputs
------
weekstring: the saturday in question
indict: a dictionary with keys being saturdays, and values being the billboard html associated
with those saturdays
   
Returns
-------
A list of tuples. Each tuple has two elements. The first element is the ranking. The
second is a dictionary songdict with keys-
    ranking: ranking of song. same as first element of the tuple
    title: title of song


Notes
-----
See description above.
"""
#your code here
def get_weekly(weekstring,indict):
  song_list=[]
  soup=BeautifulSoup(indict[weekstring],'html.parser')
  parent=soup.find_all('li',attrs={'class':'chart-list__element display--flex'})#Parent class for the table rows
  for p in parent:
   rank=p.findChild('span',attrs={'class':'chart-element__rank__number'}).text.strip()#Getting each row rank
   song=p.findChild('span',attrs={'class':'chart-element__information__song text--truncate color--primary'}).text.strip()#Getting the song name
   songdict={}
   songdict['ranking']=int(rank)
   songdict['title']=song
   song_list.append((int(rank),songdict))#appending data to the songdict dictionary
  return song_list
````

````python
"""
Function
--------
get_year

Inputs
------
weekstring: 'end_year'
indict: the value corresponding to 'end_year' in the dictionary satdict
   
Returns
-------
A list of tuples. Each tuple has two elements. The first element is the ranking. The
second is a dictionary songdict with keys-
    ranking: ranking of song. same as first element of the tuple
    title: title of song


Notes
-----
See description above.
"""
#your code here
#similar function for the getting year end details
def get_year(weekstring,indict):
  song_list=[]
  soup=BeautifulSoup(indict[weekstring],'html.parser')
  parent=soup.find_all('article',attrs={'class':'ye-chart-item'})
  for p in parent:
   rank=p.findChild('div',attrs={'class':'ye-chart-item__rank'}).text.strip()
   song=p.findChild('div',attrs={'class':'ye-chart-item__title'}).text.strip()
   songdict={}
   songdict['ranking']=int(rank)
   songdict['title']=song
   song_list.append((int(rank),songdict))
  return song_list
````

#### 2.2.3 Pre-Processing: Songs and the weeks they ranked in 

Here, write another function `get_for_title(weektups)`.

* It should take in the tuples created above.
* It should return a dictionary `titles`
  * The *key* should the title of the song
  * The *value* should be a list of tuples: (`weekindex`, `ranking`). 
    * `weekindex` should starts at 1 for the first week of the year and ends at 52 for the last week of the year
    * `ranking` is the ranking achieved that week. 

So the dictionary `titles` should look something like this:
```
{"'Tis The Damn Season": [(52, '39')],
 "(There's No Place Like) Home For The Holidays": [(1, '37'),
  (50, '47'),
  (51, '37'),
  (52, '43')],
 '10,000 Hours': [(1, '17'),
  (2, '9'),
  (3, '5'),
  (4, '7')],
...}
```

````python
"""
Function
--------
get_for_title

Inputs
------
weektups: zip of the weekinfo together with the saturdays dates
   
Returns
-------
titles: A dictionary keyed by title of the song
and with values as a list of tuples (weekindex,ranking) to indicate every ranking for that particular song corresponding to weekindex ranging from 1 to 52

Notes
-----
See description above.
"""

def get_for_title(weektups):
    titles={}
    #your code here 
    week_list=list(weektups)
    for week,songlist in enumerate(week_list):
      for rank in songlist[1]:
        try:
          titles[rank[1]['title']].append((week+1,int(rank[1]['ranking'])))#Appending the week and rank tupple if dictionary entry exists
        except KeyError:#If error
          titles[rank[1]['title']]=[(week+1,int(rank[1]['ranking']))]#Creating the new entry in the dictionary
    return titles
````

#### 2.2.4 Pre-Processing: Songs and their rankings as seen throughout the year

Finally, let's create a dictionary `tdict` keyed by the lower-cased title where we store the rank means (`rankmean`), rank standard deviations (`rankstd`) and rank count, the number of times(weeks) it appeared in the rankings in 2020, (`ranklen`), for a song.

Therefore, `tdict` should look something like this:
```
{"'tis the damn season": {'ranks': (39.0, nan, 1)},
 '10,000 hours': {'ranks': (19.31578947368421, 12.614157088934936, 19)},
```

````python
tdict={}
for title in titles.keys():
    #your code here 
    weeks,ranks=zip(*titles[title])#getting the weeks and ranks from the tuples in the dictionary
    tdict[title]={'ranks':(np.mean(ranks),np.std(ranks,ddof=1),len(weeks))};#Using ddof=1 argument to get the unbiased sample standard deviation. More on this later.
````

Do you observe the NAN values? Why does standard deviation have NAN values?



ANS: <br>Nan values occur only if we use `numpy.std` with `ddof` argument =1. By default, `ddof` argument is set to 0 in `numpy.std`. Conversely, in `pandas.std`, the ddof argument is set to 1 by default. `ddof` argument indicates which formula we are using to calculate the standard deviation of a data. If the data is indicative of the whole population, the formula for standard deviation is: <br>
$\sigma_{population}=\sqrt{\frac{\sum{(X-\mu)}^2}{n}}$ <br>
where:

*   *X= value in data*
*   *$\mu$= Population mean*
*   *n=Total number of Observations*


But if we are considering a sample from the population the formula changes to: <br>
$\sigma_{sample}=\sqrt{\frac{\sum{(x-\overline{x})}^2}{n-1}}$ <br>
The reason for this change is because using n as the total number of observation causes us to **under-estimate** the sample standard deviation compared with the population standard deviation and it is called biased standard deviation. On the other hand, using n-2 as the total number of observation causes us to over-estimate the standard deviation. However, using n-1 observation gives us a quite accurate estimate of the standard deviation for the sample. The `ddof`=1 argument specifies that we want unbiased standard deviation, but for samples with sample size 1, this causes the denominator to be 0 and the standard deviation to be infinity which basically tells us that the sample is not at all indicative of the population.
